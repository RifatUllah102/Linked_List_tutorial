```python
class Node:
  def __init__(self, Data = None, Next = None):
    self.Data = Data
    self.Next = Next

class LinkedList:
  def __init__(self):
    self.Head = None

  def insert_at_begining(self, Data):
    node = Node(Data, self.Head)
    self.Head = node

  def insert_at_end(self, Data):
    if self.Head == None:
      node = Node(Data, self.Head)
      self.Head = node
      return
    itr = self.Head
    while itr.Next:
      itr = itr.Next
    itr.Next = Node(Data, None)

  def insert_Value_list(self, data_list):
    self.Head = None
    for i in data_list:
      self.insert_at_end(i)
    
  def get_length(self):
    count = 0
    itr = self.Head
    while itr:
      count = count + 1
      itr = itr.Next
    # print(f"Length of the Linked List is {count}")
    return count

  def Print(self):
    if self.Head == None:
      print("Linked List is Empty.")
      return
    itr = self.Head
    datalist = ""
    while itr:
      datalist += str(itr.Data) + '-->'
      itr = itr.Next
    print(datalist)

  def remove_at(self, index):
    length = self.get_length()
    if index not in range(0, length):
      print("Index is out of range.")
      return
    if index == 0:
      itr = self.Head
      self.Head = itr.Next
      return
    itr = self.Head
    cnt = 0
    while itr:
      if cnt == index - 1:
        itr.Next = itr.Next.Next
        break
      cnt = cnt + 1
      itr = itr.Next
  
  def insert_at(self, index, Data):
    length = self.get_length()
    if index not in range(0, length):
      print("Index is out of range.")
      return
    if index == 0:
      self.insert_at_begining(Data)
      return
    itr = self.Head
    cnt = 0
    while itr:
      if cnt == index - 1:
        itr.Next = Node(Data, itr.Next)
        break
      cnt = cnt + 1
      itr = itr.Next
  
  def remove_by_value(self, Data):
    if self.Head == None:
      print("Linked List is Empty.")
      return
    itr = self.Head
    cnt = 0
    while itr:
      if itr.Next != None and itr.Next.Data == Data:
        itr.Next = itr.Next.Next
        cnt = 1
        break
      itr = itr.Next
    
    if cnt == 0:
      print(f"{Data} does not exist in the Linked List.")

  def insert_after_value(self, data_after, data_to_insert):
    # Search for first occurance of data_after value in linked list
    # Now insert data_to_insert after data_after node
    if self.Head == None:
      print("Linked List is Empty.")
      return
    itr = self.Head
    cnt = 0
    while itr:
      if itr.Data == data_after:
        cnt = 1
        itr.Next = Node(data_to_insert, itr.Next)
        break
      itr = itr.Next
    if cnt == 0:
      print(f"{data_after} does not exist in the Linked List.")


if __name__ == "__main__":
  ll = LinkedList()
  ll.insert_at_begining(5)
  ll.insert_at_begining(4)
  ll.insert_at_begining(3)
  ll.insert_at_begining(2)
  ll.insert_at_begining(1)
  ll.insert_at_end(6)
  ll.insert_at_end(7)
  ll.Print()
  ll.insert_Value_list(["Apple", "Banana", "Jackfruit", "Mango"])
  ll.Print()
  # ll.get_length()
  ll.insert_at(0, "Pineapple")
  ll.Print()
  ll.insert_at(3, "Guava")
  ll.Print()
  ll.remove_at(7)
  ll.remove_at(0)
  ll.Print()
  ll.remove_by_value("Guava")
  ll.Print()
  ll.remove_by_value("Rifat")
  ll.Print()
  ll.insert_after_value("Banana","Berry") # insert Berry after Banana
  ll.Print()
  ll.insert_after_value("banana","Berry")
```

```python
class Node:
    def __init__(self, data=None, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def print_forward(self):
        if self.head is None:
            print("Linked list is empty")
            return

        itr = self.head
        llstr = ''
        while itr:
            llstr += str(itr.data) + ' --> '
            itr = itr.next
        print(llstr)

    def print_backward(self):
        if self.head is None:
            print("Linked list is empty")
            return

        last_node = self.get_last_node()
        itr = last_node
        llstr = ''
        while itr:
            llstr += itr.data + '-->'
            itr = itr.prev
        print("Link list in reverse: ", llstr)

    def get_last_node(self):
        itr = self.head
        while itr.next:
            itr = itr.next

        return itr

    def get_length(self):
        count = 0
        itr = self.head
        while itr:
            count+=1
            itr = itr.next

        return count

    def insert_at_begining(self, data):
        if self.head == None:
            node = Node(data, self.head, None)
            self.head = node
        else:
            node = Node(data, self.head, None)
            self.head.prev = node
            self.head = node

    def insert_at_end(self, data):
        if self.head is None:
            self.head = Node(data, None, None)
            return

        itr = self.head

        while itr.next:
            itr = itr.next

        itr.next = Node(data, None, itr)

    def insert_at(self, index, data):
        if index<0 or index>self.get_length():
            raise Exception("Invalid Index")

        if index==0:
            self.insert_at_begining(data)
            return

        count = 0
        itr = self.head
        while itr:
            if count == index - 1:
                node = Node(data, itr.next, itr)
                if node.next:
                    node.next.prev = node
                itr.next = node
                break

            itr = itr.next
            count += 1

    def remove_at(self, index):
        if index<0 or index>=self.get_length():
            raise Exception("Invalid Index")

        if index==0:
            self.head = self.head.next
            self.head.prev = None
            return

        count = 0
        itr = self.head
        while itr:
            if count == index:
                itr.prev.next = itr.next
                if itr.next:
                    itr.next.prev = itr.prev
                break

            itr = itr.next
            count+=1

    def insert_values(self, data_list):
        self.head = None
        for data in data_list:
            self.insert_at_end(data)


if __name__ == '__main__':
    ll = DoublyLinkedList()
    ll.insert_values(["banana","mango","grapes","orange"])
    ll.print_forward()
    ll.print_backward()
    ll.insert_at_end("figs")
    ll.print_forward()
    ll.insert_at(0,"jackfruit")
    ll.print_forward()
    ll.insert_at(6,"dates")
    ll.print_forward()
    ll.insert_at(2,"kiwi")
    ll.print_forward()
```
